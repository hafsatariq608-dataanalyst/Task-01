## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd
import numpy as np

# Load the starter dataset locally
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

# Clean the GSC position gotcha (0 means no data, replace with NaN)
df['clean_position'] = df['avg_position'].replace(0, np.nan)

# Calculate key supporting numbers for our lane
total_pages = len(df)
declining_pages = df[df['trend_direction'] == 'down'].shape[0]
pct_declining = (declining_pages / total_pages) * 100
avg_ctr = df['ctr'].mean()

print("--- WEEK 1 DATA AUDIT RESULTS ---")
print(f"Total Content Items: {total_pages:,}")
print(f"Active Declining Content Items: {declining_pages:,} ({pct_declining:.2f}%)")
print(f"Average CTR across all items: {avg_ctr:.3f}%")
print(f"Items with missing position tracking (avg_position=0): {df['avg_position'].eq(0).sum():,}")
```

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

* **What I CAN claim:** We are building a highly focused, evidence-based **decision-support tool** to rank and prioritize the content review queue. This queue is sorted by a page's statistical likelihood of active organic performance decay, allowing writers to focus on high-priority opportunities first.

* **What I CANNOT claim:** We cannot claim that performing a content refresh will guaranteed immediate traffic recovery. Proving absolute causal recovery requires controlled post-update testing (A/B testing), which this observational historical dataset alone cannot verify.

* **Avoiding circular logic (Leakage):** To prevent target leakage, we are strictly building our features from raw, observable performance indicators. Downstream product decision flags (such as product-assigned scores or future flags) are entirely excluded so our model learns patterns from the real world rather than learning rules we defined ourselves.
